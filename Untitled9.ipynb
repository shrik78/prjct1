```python
!pip install transformers

from transformers import Conversation, pipeline
import random

zephyr_pipeline = pipeline("conversational", model="mistralai/Mistral-7B-v0.1", device=0)

fun_facts = [
    "Did you know that honey never spoils? Archaeologists have found pots of honey in ancient Egyptian tombs that are over 3,000 years old and still perfectly edible!",
    "Here's a fun fact: The world's oldest known recipe is for beer! It was written on a 4,000-year-old Sumerian clay tablet.",
    "Fun fact: There are more possible iterations of a game of chess than there are atoms in the known universe!",
    "Did you know that a group of flamingos is called a 'flamboyance'?",
    "Fun fact: Cows have best friends and can become stressed when they are separated from them."
]

def chat_with_zephyr():
    print("You are now chatting with Zephyr-7B-β. Type 'exit' to end the conversation.")
    conversation = Conversation()

    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            break

        conversation.add_user_input(user_input)

        zephyr_response = zephyr_pipeline(conversation)

        if random.random() < 0.2:
            fun_fact = random.choice(fun_facts)
            print("Zephyr:", fun_fact)

        print("Zephyr:", zephyr_response.generated_responses[-1])

chat_with_zephyr()
```
